# EDA and Modeling

Exploratory data analysis and step-by-step modeling for Time Series Regression (Trend + Seasonality).


```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

DATA_PATH = Path('..') / 'data' / 'train.csv'
df = pd.read_csv(DATA_PATH, parse_dates=['date'])
df.head()
```

```python
# Feature engineering (demo)
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df.groupby(['year','month']).agg({'sales':'sum'}).plot()
plt.show()
```
